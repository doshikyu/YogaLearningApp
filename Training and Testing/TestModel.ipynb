```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
import numpy as np
import cv2
import os
```

```python
test_path ='C:/Users/dasha/Desktop/Study/Project/ImagesDataset/Advanced/test'
```

```python
model = keras.models.load_model('C:/Users/dasha/Desktop/Study/Project/advancedModel1.h5')
model.load_weights('C:/Users/dasha/Desktop/Study/Project/advancedModel_weights1.h5')
```

```python
for i in os.listdir(test_path):
    img = image.load_img(test_path+'//'+i, target_size=(300,300))
    plt.imshow(img)
    plt.show()
    
    X=image.img_to_array(img)
    X=np.expand_dims(X,axis=0)
    images=np.vstack([X])
   
    val = model.predict(images, batch_size=5)
    
    print(val)
    
    #{'childspose': 0, 'downdog': 1, 'easypose': 2}
    #{'cobra': 0, 'plank': 1, 'sideplank': 2}
  
```
