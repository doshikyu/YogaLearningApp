```python
import tkinter as tk
from tkinter import ttk
import customtkinter
from tkinter import *
import numpy as np
import cv2
import os
from PIL import Image, ImageTk
customtkinter.set_appearance_mode("Dark")
customtkinter.set_default_color_theme("green")
```

```python
WIDTH = 1100
HEIGHT = 600
APP_NAME = "CustomTkinter example_background_image.py"
```

```python
class App(customtkinter.CTk):
    def __init__(self, *args, **kwargs):
        customtkinter.CTk.__init__(self, *args, **kwargs)
        self.title(APP_NAME)
        self.geometry(f"{WIDTH}x{HEIGHT}")
        self.protocol("WM_DELETE_WINDOW", self.on_closing)    
        self.resizable(False,False)
        
        container = customtkinter.CTkFrame(master=self)
        container.pack(fill="both", expand=True)
        container.grid_rowconfigure(0, weight = 1)
        container.grid_columnconfigure(0, weight = 1)
        
        #switch between pages
        self.frames = {} 
        for F in (StartPage, PracticePage):
            frame = F(container, self)
            self.frames[F] = frame
            frame.grid(row = 0, column = 0, sticky ="nsew")
            
        self.show_frame(StartPage)
  
    def show_frame(self, cont):
        frame = self.frames[cont]
        frame.tkraise() #The top frame will be visible (risen)
    
    def on_closing(self, event=0):
        self.destroy()
        
    def change_appearance_mode(self, new_appearance_mode):
        customtkinter.set_appearance_mode(new_appearance_mode)
        
    def font_style():
        label.config(font=('Helvetica bold', 26))
        
```

```python
class StartPage(customtkinter.CTkFrame):

    def __init__(self, parent, controller):
        customtkinter.CTkFrame.__init__(self, parent)
        self.frame_left = customtkinter.CTkFrame(master=self, width=WIDTH-20, height = HEIGHT-20, corner_radius=20)
        self.frame_left.pack(fill="both", expand=True, padx=20, pady=20)
        
        #grid
        self.frame_left.grid_columnconfigure(0, weight=1)
        self.frame_left.grid_columnconfigure(1, weight=1)
        self.frame_left.grid_columnconfigure(2, weight=1)
        self.frame_left.grid_rowconfigure(0, weight=5)
        #self.frame_left.grid_rowconfigure(1, weight=1)
        self.frame_left.grid_rowconfigure(2, weight=2)
        
        #labels
        label = customtkinter.CTkLabel(master=self.frame_left, text ="YOGA APP", text_font=("Impact", 60))
        label.grid(row = 0, column = 0, columnspan=3, padx = 10, pady = 10)
        label = customtkinter.CTkLabel(master=self.frame_left, text ="Choose your level")
        label.grid(row = 1, column = 1, padx = 10, pady = 10)
        
        
        #buttons
        self.button_1 = customtkinter.CTkButton(master=self.frame_left,text="Beginner", command = lambda : controller.show_frame(PracticePage))
        self.button_1.grid(row=2, column=0, pady=20, padx=30)
        self.button_2 = customtkinter.CTkButton(master=self.frame_left,text="Intermediate", command = lambda : controller.show_frame(PracticePage))
        self.button_2.grid(row=2, column=1, pady=20, padx=30)
        self.button_3 = customtkinter.CTkButton(master=self.frame_left,text="Advanced", command = lambda : controller.show_frame(PracticePage))
        self.button_3.grid(row=2, column=2, pady=20, padx=30)
        
      
```

```python
class PracticePage(customtkinter.CTkFrame):
    def __init__(self, parent, controller):
        customtkinter.CTkFrame.__init__(self, parent)
        self.grid_columnconfigure(1, weight=1)
        self.grid_rowconfigure(0, weight=1)

        self.frame_left = customtkinter.CTkFrame(master=self,
                                                 width=180,
                                                 corner_radius=0)        
        self.frame_left.grid(row=0, column=0, rowspan=2, sticky="nswe")

        self.frame_right = customtkinter.CTkFrame(master=self)
        self.frame_right.grid(row=0, column=1, sticky="nswe", padx=20, pady=20)

        #grid
        self.frame_left.grid_rowconfigure(0, minsize=10)   # empty row with minsize as spacing
        self.frame_left.grid_rowconfigure(5, weight=1)  # empty row as spacing
        self.frame_left.grid_rowconfigure(8, minsize=20)    # empty row with minsize as spacing
        self.frame_left.grid_rowconfigure(11, minsize=10)  # empty row with minsize as spacing
        
        self.frame_right.grid_rowconfigure(0, weight=2)
        self.frame_right.grid_rowconfigure(1, weight=1)
        self.frame_right.grid_columnconfigure(0, weight=1)
        self.frame_right.grid_columnconfigure(1, weight=1)

        #left screen
        self.button_2 = customtkinter.CTkButton(master=self.frame_left,
                                                text="Main Menu",command = lambda : controller.show_frame(StartPage))
        self.button_2.grid(row=8, column=0, pady=10, padx=20)
        
        self.button_2 = customtkinter.CTkButton(master=self.frame_left,
                                                text="Go to Practice")
        self.button_2.grid(row=7, column=0, pady=10, padx=20)

        self.label_level = customtkinter.CTkLabel(master=self.frame_left, text="Pose Select:", text_font=("Impact", 15) )
        self.label_level.grid(row=2, column=0, pady=10, padx=20, sticky="w")
        
        self.optionmenu_2 = customtkinter.CTkOptionMenu(master=self.frame_left,values=["Choose","DownDog.jpg", "Tree.jpg", "Warrior.JPEG"], command= self.change)
        self.optionmenu_2.grid(row=3, column=0, pady=10, padx=20, sticky="w")
        
        #right screen
        self.label_image = customtkinter.CTkLabel(master = self.frame_right, text = "Select the Pose", image=None)
        self.label_image.grid(row = 0, column=0, columnspan=1, pady=20, padx=10, sticky="")
        
        self.label_level = customtkinter.CTkLabel(master=self.frame_right, text_font=("Bahnschrift", 8), text="Downward dog is an effective yoga pose that can help relieve chronic pains caused by sitting, tight hamstrings. \n1. Get down on all fours. Start on the floor with your hands shoulder-width apart, with your shoulders above your wrists. Align your inner shoulders with your index fingers. Your hips should be above or slightly before your knees. \n2. Lift your knees. Next, tuck your toes against the mat or ground, using that leverage to extend your legs and lift both knees into the air. Your body should now resemble an upside-down “V” shape. \n3. Extend. Extend and lengthen your spine, simultaneously pressing through the palms of your hands and balls of your feet. Pull your pelvis up toward the ceiling, using the triceps in your upper arms to help stabilize your form. Some yoga teachers suggest drawing your shoulder blades down your back, while others prefer externally rotating the joints to support the body instead. \n4. Hold and release. Hold your body in position, making sure to breathe properly. Slowly bring your knees back to the floor to release this asana, or use it to transition into another pose.")
        self.label_level.grid(row=1, column=0 , sticky="", pady=10, padx=20)
        
    
       
    def change(self,*args):
        for arg in args:
            img = cv2.imread(arg)
            width = 350
            height = 450
            dim = (width, height)
            img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
            blue,green,red = cv2.split(img)
            img = cv2.merge((red,green,blue))
            im = Image.fromarray(img)
            imgtk = ImageTk.PhotoImage(image=im)
            self.label_image.configure(image = imgtk)
            self.label_image.image = imgtk
    
       
```

```python
app = App()
app.mainloop()
```
